```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the DataFrame
df = pd.read_csv('understat.com.csv')

# Group by 'position' and calculate the mean xG for each position
average_xG_by_position = df.groupby('position')['xG'].mean().reset_index()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(range(len(average_xG_by_position)), average_xG_by_position['xG'], color='blue')
plt.title('Average xG by Position')
plt.xlabel('Position')
plt.ylabel('Average xG')
plt.xticks(range(len(average_xG_by_position)), average_xG_by_position['position'], rotation='vertical')
plt.grid(True)
plt.show()
```
